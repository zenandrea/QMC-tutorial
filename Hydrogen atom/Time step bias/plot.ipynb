```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
El = np.loadtxt('el.dat')
Et = np.loadtxt('et.dat')
```

```python
El
```

```python
fig = plt.figure(figsize=(4,4))
ax1 = plt.gca()

ax1.set_xlim(0,0.22)
ax2 = ax1.inset_axes([1.3,0.1,1,0.8])
ax2.set_ylim(-0.5006,-0.4997)
ax2.set_xlim(0,0.013)

for ax in ax1,ax2:
    ax.axhline(-0.5,c='green',label='Exact')
    ax.errorbar(El[:,0],El[:,1],yerr=El[:,2],label=r'$\left< E_{step} \right>$',
                fmt='ob--', mec='k', ecolor='k', capsize=3)
    ax.errorbar(Et[:,0],Et[:,1],yerr=Et[:,2],label=r'$\left< E_{T} \right>$',
                fmt='or--', mec='k', ecolor='k', capsize=3)

ax1.legend(loc=2)
ax1.indicate_inset_zoom(ax2, edgecolor="black")
ax1.set_xlabel(r'$\delta\tau$ (a.u.)', fontsize=13)
ax1.set_ylabel('Energy (a.u.)', fontsize=13)
ax1.set_title('Hydrogen atom', fontsize=13)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.savefig('fig6.png',dpi=300,bbox_inches='tight')
#plt.tight_layout()
```
