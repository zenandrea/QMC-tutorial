```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
def func(c):
    return x1-x2+y1-y2+c*0.4**(0.5)*(y1**2-y2**2)
```

```python
plt.clf()
ax = plt.gca()
ax.set_xlim([0, 2])
ax.set_ylim([-1, 1])
ax.set_box_aspect(1)
#plt.title('Trial nodal surfaces')
plt.xlabel('$x_1$', fontsize=14)
plt.ylabel('$y_1$', fontsize=14)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.grid()
x2=1
y2=0
delta = 0.025
x1range = np.arange(0, 2, delta)
y1range = np.arange(-1, 1, delta)
x1, y1 = np.meshgrid(x1range, y1range)
cs0=plt.contour(x1, y1, func(0), [0], linewidths=2.5, colors='black')
cs0_5=plt.contour(x1, y1, func(0.5), [0], colors='red')
cs1_5=plt.contour(x1, y1, func(1.5), [0], colors='blue')
cs_1=plt.contour(x1, y1, func(-1), [0], colors='green')
cs_2=plt.contour(x1, y1, func(-2), [0], colors='gold')
#plt.plot(np.linspace(-1,1,400), func(1,0,np.linspace(-1,1,400),0),linewidth=3)
#plt.plot(np.linspace(-1,1,400), func(1,0,np.linspace(-1,1,400),0.5))
#plt.plot(np.linspace(-1,1,400), func(1,0,np.linspace(-1,1,400),1.5))
#plt.plot(np.linspace(-1,1,400), func(1,0,np.linspace(-1,1,400),-1))
#plt.plot(np.linspace(-1,1,400), func(1,0,np.linspace(-1,1,400),-2))
plt.plot(1,0, color="white", marker="o",  markeredgecolor="black")
h0,_ = cs0.legend_elements()
h0_5,_ = cs0_5.legend_elements()
h1_5,_ = cs1_5.legend_elements()
h_1,_ = cs_1.legend_elements()
h_2,_ = cs_2.legend_elements()
ax.legend([h1_5[0],h0_5[0],h0[0],h_1[0],h_2[0]], ['c = +1.5','c = +0.5','c = 0.0','c = -1.0','c = -2.0'])
plt.show()
```
