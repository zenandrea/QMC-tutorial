```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes, mark_inset
%matplotlib widget
```

```python
# Leggi i dati dal file di testo
data = np.loadtxt('C:/Users/Alfonso/Downloads/Fisica Universit√†/Tesi magistrale/Codice-qmc-C/Harmonic/Harm Population error_tau0.05.txt')

# Estrai le colonne
pop = data[:,0]
energy = data[:,1]/1
errorbar = data[:,2]/1

# Valore costante per E_0
E_0 = 0.5
```

```python
plt.clf()
# Crea la figura e gli assi
fig, ax = plt.subplots()

# Plotta i dati principali
ax.errorbar(pop, energy, errorbar, label=r'$\left< E_{T} \right>$', fmt='ko--', mec='k', ecolor='k', markersize=3, capsize=3, lw=0.7)
#ax.errorbar(pop, energy, errorbar, fmt='--', c='gray')
ax.set_ylim(0.499, 0.508)
ax.axhline(y=E_0, color='green', linestyle='-', linewidth=1, label='Exact')
ax.set_xscale('log')

# Aggiungi etichette e legende
ax.set_xlabel('Population target $N_{target}$', fontsize=14)
ax.set_ylabel('Energy (a.u.)', fontsize=14)
ax.legend(loc='upper right', fontsize=10)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
#ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
#fig.tight_layout()
#ax_inset.legend(loc='upper right')

# Mostra la figura
plt.show()
```
