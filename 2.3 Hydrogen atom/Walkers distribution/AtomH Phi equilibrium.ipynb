```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes, mark_inset
%matplotlib widget
```

```python
# Read data from text file
data_eq = np.loadtxt('f_onda.txt')
data = np.loadtxt('f_onda_equil.txt')

# Extract columns
r = data_eq[:,0]
phi = data_eq[:,1]
phi1 = data[(33*800):(34*800),1]
phi3 = data[(100*800):(101*800),1]
```

```python
phi1
```

```python
def func(x):
    return 4*np.sqrt(np.pi)*x**2*np.exp(-x)
```

```python
plt.clf()
# Create the figure and axes
fig, ax = plt.subplots()
ax.tick_params(axis='both', direction='in', length=7, top=True, bottom=True, left=True, right=True)

# Plot
plt.plot(r, phi, 'g-', label=r'$F_{eq}(r)=\langle F (r) \rangle_{\tau}, 20<\tau<400$ a.u.', markersize=2, lw=1.6)
plt.plot(r, phi1, 'k-', label=r'$F (r), \tau=1$ a.u.', markersize=2, lw=1)
plt.plot(r, phi3, 'b-', label=r'$F (r), \tau=3$ a.u.', markersize=2, lw=1)
ax.set_ylim(bottom=0)
ax.set_xlim(0, 10)
#ax.set_xscale('log')
plt.plot(np.linspace(0,13,500), func(np.linspace(0,13,500)), '--', label=r'$F_{0}(r)$', color='red', dashes=[6, 4], lw=3)

# Add labels and legends
ax.set_xlabel('$r$ (a.u.)', fontsize=14)
ax.set_ylabel('Radial distribution $F(r)$', fontsize=14)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
#ax.legend(loc='upper right')
ax.legend(bbox_to_anchor=(0.95, 0.95), loc='upper right', fontsize=10)
#fig.tight_layout()

# Show figure
plt.show()
```
