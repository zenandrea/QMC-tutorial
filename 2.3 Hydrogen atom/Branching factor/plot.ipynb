```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes, mark_inset
%matplotlib widget
```

```python
# Read data from text file
m_0125 = np.loadtxt('0.00125_m.dat')
m_2000 = np.loadtxt('0.02_m.dat')
Ntarget = 10**6
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, layout='constrained', figsize=(10.5,3.5))

#ax1.tick_params(axis='both', direction='in', length=7, top=True, bottom=True, left=True, right=True)

# Plot
ax1.plot(m_0125[:,0], m_0125[:,1], 'k.', markersize=2, lw=0.7)
ax1.set_ylim(0.0, 20)
ax1.set_xlim(0, 300)


# Add labels and legends
ax1.tick_params(axis='both', labelsize=13)
#ax1.legend(bbox_to_anchor=(0.97, 0.05), loc='lower right', fontsize=12)
ax1.set_title(r'$\delta\tau=0.00125$ a.u.', fontsize=14, loc='center')
ax1.set_xlabel('Instances with m>2', fontsize=14)
ax1.set_ylabel('m', fontsize=14)

#ax2.tick_params(axis='both', direction='in', length=7, top=True, bottom=True, left=True, right=True)

# Plot
ax2.plot(m_2000[:,0], m_2000[:,1], 'k.', markersize=2, lw=0.7)
#ax2.set_ylim(bottom=0.1)
ax2.set_xlim(0, 361885)
ax2.axhline(y=Ntarget, color='red', label=r'$N_{target}$', linestyle='-', linewidth=1.5)
ax2.set_yscale('log')
ax2.ticklabel_format(axis='x',style='scientific', scilimits=(0, 0), useMathText=True)

# Add labels and legends
ax2.tick_params(axis='both', labelsize=13)
ax2.legend(bbox_to_anchor=(0.97, 0.95), loc='upper right', fontsize=12)
ax2.set_title(r'$\delta\tau=0.02$ a.u.', fontsize=14, loc='center')
ax2.set_xlabel('Instances with m>2', fontsize=14)
ax2.set_ylabel('m', fontsize=14)

plt.show()
```
