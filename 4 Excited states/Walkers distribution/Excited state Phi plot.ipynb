```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes, mark_inset
%matplotlib widget
```

```python
# Read data from text file
data = np.loadtxt('Phi tau=0')
```

```python
# Extract columns
x = data[0:2200,0]
phi0 = data[0:2200,1]
```

```python
phi0
```

```python
len(phi0)
```

```python
# exact distribution
def func(x):
    return np.sqrt((np.sqrt(2)*0.4**(0.25)*x*np.exp(-x**2*0.5)*np.pi**(-0.25))**2)
```

```python
plt.clf()
# Create the figure and axes
fig, ax1 = plt.subplots()
ax1.tick_params(axis='both', direction='in', length=7, top=True, bottom=True, left=True, right=True)

# Plot
plt.plot(x*np.sqrt(0.4), phi0, 'k-', label=r'$|\phi_1|, \tau=0$ a.u.', markersize=2, lw=1)
#plt.plot(r, phi0, 'k-', label=r'$F (r), \tau=1$ a.u.', markersize=2, lw=1.6)
ax1.set_ylim(bottom=-0.003)
ax1.set_xlim(-4.5, 4.5)
plt.plot(np.linspace(-4.5,4.5,500), func(np.linspace(-4.5,4.5,500)), '--', label=r'$|\psi_1|$', color='red', lw=3)


# Add labels and legends
ax1.set_xlabel('$x \\sqrt{\\omega}$ (a.u.)', fontsize=14)
ax1.set_ylabel('$|\phi(x,\\tau)|$', fontsize=14)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
#ax.legend(loc='upper right')
ax1.legend(bbox_to_anchor=(0.97, 0.95), loc='upper right', fontsize=12)
#fig.tight_layout()

# Show figure
plt.show()
```

```python
plt.savefig('Phi tau=0.png', dpi=300, bbox_inches='tight')
```
