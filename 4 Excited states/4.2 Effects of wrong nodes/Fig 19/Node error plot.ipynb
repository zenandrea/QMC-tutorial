```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib widget
```

```python
def func(x, a, b):
    return a*x + b
```

```python
data = np.loadtxt('Node error data.txt')

x_node = data[:,0]*np.sqrt(0.4)
energy = data[:,1]/0.4
errorbar = data[:,2]/0.4

E_1 = 1.5
```

```python
plt.clf()
fig, ax = plt.subplots()
# best fit parameters
a=0.278535
b=0.599787

# Plot
ax.errorbar(x_node, energy, errorbar, fmt='ko', label=r'$\left< E_{T} \right>$', markersize=4, capsize=3)
plt.plot(np.linspace(0,0.14,500), func(np.linspace(0,0.14,500)/np.sqrt(0.4),a,b)/0.4, 'k--', lw=0.7)
ax.set_ylim(1.49,1.58)
ax.set_xlim(right=0.07)
ax.axhline(y=E_1, color='green', linestyle='-', linewidth=1, label='$E_1$')

ax.set_xlabel('$x_{node} \\sqrt{\\omega}$ (a.u.)', fontsize=14)
ax.set_ylabel('Energy$/\\omega$ (a.u.)', fontsize=14)
ax.legend(loc='lower right', bbox_to_anchor=(0.98, 0.15))
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)

plt.show()
```

```python
plt.savefig('Node error plot.png', dpi=300, bbox_inches='tight')
```
