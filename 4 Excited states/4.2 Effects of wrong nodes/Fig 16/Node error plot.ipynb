```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib widget
```

```python
def func(x, a, b):
    return a*x + b
```

```python
data = np.loadtxt('Node error data.txt')

x_node = data[:,0]*np.sqrt(0.4)
energy = data[:,1]/0.4
errorbar = data[:,2]/0.4

E_1 = 1.5
```

```python
plt.clf()
fig, ax = plt.subplots()
# best fit parameters
a=-0.273661
b=0.599545

# Plot
ax.errorbar(x_node, energy, errorbar, fmt='ko', label=r'$\left< E_{T} \right>$', markersize=4, capsize=3)
plt.plot(np.linspace(0,0.08,500), func(np.linspace(0,0.08,500)/np.sqrt(0.4),a,b)/0.4, 'k--', lw=0.7)
plt.plot(np.linspace(-0.08,0,500), func(np.linspace(-0.08,0,500)/np.sqrt(0.4),-a,b)/0.4, 'k--', lw=0.7)
ax.set_ylim(top=1.51)
ax.set_xlim(-0.07,0.07)
ax.axhline(y=E_1, color='green', linestyle='-', linewidth=1, label='$E_1$')
#ax.axhline(y=E_0, color='red', linestyle='-', linewidth=1, label='$E_0$')

ax.set_xlabel('$x_{node} \\sqrt{\\omega}$ (a.u.)', fontsize=14)
ax.set_ylabel('Energy$/\\omega$ (a.u.)', fontsize=14)
ax.legend(loc='upper right', bbox_to_anchor=(0.98, 0.85))
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)

plt.show()
```

```python
plt.savefig('Node error plot.png', dpi=300, bbox_inches='tight')
```
