```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes, mark_inset
%matplotlib widget
```

```python
data = np.loadtxt('Node error data.txt')

x_node = data[:,0]*np.sqrt(0.4)
energy = data[:,1]/0.4
errorbar = data[:,2]/0.4

E_1 = 1.5
E_0 = 0.5
```

```python
plt.clf()
fig, ax = plt.subplots()

# Plot
ax.errorbar(x_node, energy, errorbar, fmt='ko--', label=r'$\left< E_{T} \right>$', markersize=4, capsize=3)
ax.set_ylim(0.42, 1.58)
ax.axhline(y=E_1, color='green', linestyle='-', linewidth=1, label='$E_1$')
ax.axhline(y=E_0, color='red', linestyle='-', linewidth=1, label='$E_0$')
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)

# Inset
ax_inset = fig.add_axes([0.45, 0.45, 0.32, 0.28])
ax_inset.errorbar(x_node[:10], energy[:10], errorbar[:10], fmt='ko--', markersize=4)
ax_inset.set_xlim(-0.01, 0.07)
ax_inset.set_ylim(1.40, 1.51)
ax_inset.axhline(y=E_1, color='green', linestyle='-', linewidth=1)
ax_inset.axhline(y=E_0, color='red', linestyle='-', linewidth=1)
mark_inset(ax, ax_inset, loc1=1, loc2=3, fc="none", ec="0.6")

ax.set_xlabel('$x_{node} \\sqrt{\\omega}$ (a.u.)', fontsize=14)
ax.set_ylabel('Energy$/\\omega$ (a.u.)', fontsize=14)
ax.legend(loc='lower right', bbox_to_anchor=(0.98, 0.1))

plt.show()
```

```python
plt.savefig('Node error plot.png', dpi=300, bbox_inches='tight')
```
