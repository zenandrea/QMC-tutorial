```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes, mark_inset
%matplotlib widget
```

```python
# Read data from text file
data = np.loadtxt('f_onda_0.txt')
```

```python
x = data[0:2200,0]
phi0 = data[0:2200,1]
```

```python
phi0
```

```python
len(phi0)
```

```python
def func(x):
    return np.sqrt((np.sqrt(2)*0.4**(0.25)*x*np.exp(-x**2*0.5)*np.pi**(-0.25))**2)
xnode=0.2*np.sqrt(0.4)
```

```python
plt.clf()
# Create the figure and axes
fig, ax = plt.subplots()
ax.tick_params(axis='both', direction='in', length=7, top=True, bottom=True, left=True, right=True)

# Plot
plt.plot(x*np.sqrt(0.4), phi0, 'k-', label=r'$|\phi_1|, \tau=0$ a.u.', markersize=2, lw=1)
#plt.plot(r, phi0, 'k-', label=r'$F (r), \tau=1$ a.u.', markersize=2, lw=1.6)
ax.set_ylim(bottom=-0.003)
ax.set_xlim(-2.5, 2.5)
#ax.axhline(y=E_0, color='green', linestyle='-', linewidth=1, label='Exact')
#ax.set_xscale('log')
plt.plot(np.linspace(-4.5,4.5,500), func(np.linspace(-4.5,4.5,500)), '--', label=r'$|\psi_1|$', color='red', lw=3)
plt.vlines(xnode,0,0.35, linestyles=':', label='Trial node', lw=3)
plt.scatter(xnode,0.35, s=200, marker='x', lw=2)

# Add labels and legends
ax.set_xlabel('$x \\sqrt{\\omega}$ (a.u.)', fontsize=14)
ax.set_ylabel('$|\phi(x,\\tau)|$', fontsize=14)
#ax.legend(loc='upper right')
ax.legend(bbox_to_anchor=(0.03, 0.95), loc='upper left')
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
#fig.tight_layout()

# Show figure
plt.show()
```

```python
plt.savefig('Phi tau=0.png', dpi=300, bbox_inches='tight')
```

```python
data = np.loadtxt('f_onda.txt')
xeq = data[:,0]
phieq = data[:,1]
```

```python
plt.clf()
fig, ax = plt.subplots()
ax.tick_params(axis='both', direction='in', length=7, top=True, bottom=True, left=True, right=True)

# Plot
plt.plot(xeq*np.sqrt(0.4), phieq, 'g-', label=r'$|\phi_1|_{eq}, \tau=25610$ a.u.', markersize=2, lw=2)
#plt.plot(r, phi0, 'k-', label=r'$F (r), \tau=1$ a.u.', markersize=2, lw=1.6)
ax.set_ylim(bottom=-0.002)
ax.set_xlim(-3.5, 3.5)
plt.plot(np.linspace(-4.5,4.5,500), func(np.linspace(-4.5,4.5,500)), '--', label=r'$|\psi_1|$', color='red', lw=3)

ax.set_xlabel('$x \\sqrt{\\omega}$ (a.u.)', fontsize=14)
ax.set_ylabel('$|\phi(x,\\tau)|$', fontsize=14)
#ax.legend(loc='upper right')
ax.legend(bbox_to_anchor=(0.03, 0.95), loc='upper left')
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)

plt.show()
```

```python
plt.savefig('Phi equilibrium.png', dpi=300, bbox_inches='tight')
```
