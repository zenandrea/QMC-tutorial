```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes, mark_inset
%matplotlib widget
```

```python
# Read data from text file
phi_0 = np.loadtxt('phi_tau=0.dat')
phi_0_5 = np.loadtxt('phi_tau=0.5.dat')
phi_1 = np.loadtxt('phi_tau=1.dat')
phi_2 = np.loadtxt('phi_tau=2.dat')
psi0 = np.loadtxt('psi0.dat')
```

```python
plt.clf()
# Create the figure and axes
fig, ax = plt.subplots()
ax.tick_params(axis='both', direction='in', length=7, top=True, bottom=True, left=True, right=True)

# Plot
plt.plot(phi_0[:,0], phi_0[:,1], 'k-', label=r'$\tau=0$ a.u.', markersize=2, lw=0.8)
plt.plot(phi_0_5[:,0], phi_0_5[:,1], '-', color='aqua', label=r'$\tau=0.5$ a.u.', markersize=2, lw=1.2)
plt.plot(phi_1[:,0], phi_1[:,1], 'b-', label=r'$\tau=1$ a.u.', markersize=2, lw=1.2)
plt.plot(phi_2[:,0], phi_2[:,1], 'g-', label=r'$\tau=2$ a.u.', markersize=2, lw=1.2)
plt.plot(psi0[:,0], psi0[:,1], 'r--', label=r'$\Psi_0$', markersize=2, dashes=[6, 4], lw=1.5)

ax.set_ylim(0, 0.6)
ax.set_xlim(-4, 4)

# Add labels and legends
ax.set_xlabel('$x$ (a.u.)', fontsize=14)
ax.set_ylabel(r'$\phi(x,\tau)$', fontsize=14)
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
#ax.legend(loc='upper right')
ax.legend(bbox_to_anchor=(0.97, 0.95), loc='upper right', fontsize=11)
#fig.tight_layout()

# Mostra la figura
plt.show()
```
