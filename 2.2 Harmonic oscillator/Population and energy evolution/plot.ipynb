```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes, mark_inset
%matplotlib widget
```

```python
# Read data from text file
el = np.loadtxt('el.dat')
et = np.loadtxt('et.dat')
pop = np.loadtxt('pop.dat')
Ntarget = 10**5
E_0 = 0.5
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, layout='constrained', figsize=(10.5,3.5))

#ax1.tick_params(axis='both', direction='in', length=7, top=True, bottom=True, left=True, right=True)

# Plot
ax1.plot(pop[:,0], pop[:,1]/10**5, 'k-', markersize=2, label='Population', lw=0.7)
#ax1.set_ylim(bottom=-0.003)
ax1.set_xlim(0, 10)
ax1.axhline(y=Ntarget/10**5, color='red', label=r'$N_{target}$', linestyle='-', linewidth=1.5)


# Add labels and legends
ax1.tick_params(axis='both', labelsize=13)
ax1.legend(bbox_to_anchor=(0.97, 0.05), loc='lower right', fontsize=12)
#ax1.set_title('(a)', fontsize=14, loc='left')
ax1.set_xlabel(r'$\tau$ (a.u.)', fontsize=14)
ax1.set_ylabel(r'Walker population ($\times 10^5$)', fontsize=14)

#ax2.tick_params(axis='both', direction='in', length=7, top=True, bottom=True, left=True, right=True)

# Plot
ax2.plot(et[:,0], et[:,1], 'k-', label=r'$E_T$', markersize=2, lw=0.7)
ax2.plot(el[:,0], el[:,1], '-', color='limegreen', label=r'$E_{step}$', markersize=2, lw=1)
ax2.set_ylim(bottom=0.1)
ax2.set_xlim(0, 10)
ax2.axhline(y=E_0, color='red', linestyle='-', linewidth=1.5, label=r'$E_0=0.5$ a.u.')

# Add labels and legends
ax2.tick_params(axis='both', labelsize=13)
ax2.legend(bbox_to_anchor=(0.97, 0.05), loc='lower right', fontsize=12)
#ax2.set_title('(b)', fontsize=14, loc='left')
ax2.set_xlabel(r'$\tau$ (a.u.)', fontsize=14)
ax2.set_ylabel('Energy (a.u.)', fontsize=14)

plt.show()
```
